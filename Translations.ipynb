```python
from transformers import pipeline
import csv

# 1. 載入翻譯模型
fr_to_en_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")  # 法語到英文
en_to_zh_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-zh")  # 英文到中文

# 2. 讀取法文原文文件
input_file = "original_texts.txt"  # 法文原文文件
output_text_file = "translations_results.txt"  # 翻譯結果保存到文本
output_csv_file = "translations_results.csv"  # 翻譯結果保存到 CSV

# 讀取文件中的每行法文文本
with open(input_file, "r", encoding="utf-8") as file:
    french_texts = file.readlines()

# 去掉多餘的空行和換行符
french_texts = [line.strip() for line in french_texts if line.strip()]

# 3. 定義後處理函數
def post_process_translation(translation):
    # 修正翻譯中的常見語法和格式問題
    translation = translation.replace("：", "： ")  # 確保冒號後有空格
    translation = translation.replace(" 。", "。")  # 修正句號前多餘空格
    translation = translation.replace(" ！", "！")  # 修正感嘆號前多餘空格
    return translation

# 4. 逐段翻譯，生成多版本翻譯
translations = []

for french_text in french_texts:
    print(f"翻譯中：{french_text}")

    # 第一步：法語到英文翻譯
    standard_english = fr_to_en_translator(french_text)[0]['translation_text']
    poetic_english = fr_to_en_translator("Poetic style: " + french_text)[0]['translation_text']
    colloquial_english = fr_to_en_translator("Colloquial style: " + french_text)[0]['translation_text']

    # 第二步：英文到中文翻譯
    standard_chinese = en_to_zh_translator(standard_english)[0]['translation_text']
    poetic_chinese = en_to_zh_translator(poetic_english)[0]['translation_text']
    colloquial_chinese = en_to_zh_translator(colloquial_english)[0]['translation_text']

    # 後處理翻譯
    standard_chinese = post_process_translation(standard_chinese)
    poetic_chinese = post_process_translation(poetic_chinese)
    colloquial_chinese = post_process_translation(colloquial_chinese)

    # 保存翻譯結果到字典
    translations.append({
        "原文": french_text,
        "標準翻譯（英文）": standard_english,
        "詩意翻譯（英文）": poetic_english,
        "口語化翻譯（英文）": colloquial_english,
        "標準翻譯（中文）": standard_chinese,
        "詩意翻譯（中文）": poetic_chinese,
        "口語化翻譯（中文）": colloquial_chinese,
    })

# 5. 保存翻譯結果到文本文件
with open(output_text_file, "w", encoding="utf-8") as file:
    for t in translations:
        file.write(f"原文：{t['原文']}\n")
        file.write(f"標準翻譯（英文）：{t['標準翻譯（英文）']}\n")
        file.write(f"詩意翻譯（英文）：{t['詩意翻譯（英文）']}\n")
        file.write(f"口語化翻譯（英文）：{t['口語化翻譯（英文）']}\n")
        file.write(f"標準翻譯（中文）：{t['標準翻譯（中文）']}\n")
        file.write(f"詩意翻譯（中文）：{t['詩意翻譯（中文）']}\n")
        file.write(f"口語化翻譯（中文）：{t['口語化翻譯（中文）']}\n\n")
print(f"翻譯結果已保存到文本文件：{output_text_file}")

# 6. 保存翻譯結果到 CSV 文件
with open(output_csv_file, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=[
        "原文", "標準翻譯（英文）", "詩意翻譯（英文）", "口語化翻譯（英文）",
        "標準翻譯（中文）", "詩意翻譯（中文）", "口語化翻譯（中文）"
    ])
    writer.writeheader()
    for t in translations:
        writer.writerow(t)
print(f"翻譯結果已保存到 CSV 文件：{output_csv_file}")
```
